```python
#per ogni y ho una funzione polinomiale => matrice

import numpy as np
import matplotlib.pyplot as plt

X = np.random.randint(0, 200, (250,1))

Y = X**2 + np.random.randint(900, 10000, (250,1))

plt.scatter(X, Y)
plt.show()
```

```python
weight = np.array([0,0,0], float)
feature = np.array([np.ones((250,1)), X, X**2])
partial_grad = np.array([0,0,0], float)

alpha = 0.1
epochs = 10000

for _ in range(epochs):
    for i in range(len(feature)):
        pred = weight[2] * feature[2] + weight[1] * feature[1] + weight[0] * feature[1]

        partial_grad[i] = sum(feature[i] * (Y - pred))

        
    
    normalized_grad = np.sqrt(sum(partial_grad**2))
        
    for i in range(len(feature)):
        weight[i] += (partial_grad[i]/normalized_grad) * alpha


print(weight) 
```

```python
plt.scatter(X, weight[2] * feature[2] + weight[1] * feature[1] + weight[0] * feature[0], color="red")
plt.scatter(X, Y)
```
