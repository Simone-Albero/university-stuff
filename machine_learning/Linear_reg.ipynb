```python
import numpy as np
import matplotlib.pyplot as plt

X = np.random.randint(0, 100, (250,1))

Y = X**2 + np.random.randint(200, 2000, (250,1))

plt.scatter(X, Y)
plt.show()

```

```python
m = 0
q = 0
n = len(X)

epochs = 20000
alpha = 0.1

for _ in range(epochs):
    pred = m * X + q
    d_m = (-2) * sum(X * (Y - pred))
    d_q = (-2) * sum(Y - pred)

    normalized_grad = np.sqrt(d_m**2 + d_q**2)
    if(normalized_grad > 1):
        d_m = d_m/normalized_grad
        d_q = d_q/normalized_grad

    m -= alpha * d_m
    q -= alpha * d_q

print("grad: ", normalized_grad)
print("cost: ", np.sqrt(sum(np.square(Y - pred)))/n)
print("m:" , m , "q:" , q)
```

```python
plt.scatter(X, m * X + q, color="red")
plt.scatter(X, Y)
```
